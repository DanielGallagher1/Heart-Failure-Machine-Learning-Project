# PyCaret Model

```python
from pycaret.classification import *
s = setup(train, target = 'DEATH_EVENT', session_id = 123)
```

# Compare Models

```python
# functional API
best = compare_models()

# OOP API
best = s.compare_models()
```

```python
print(best)
```

# Analyze Model

```python
# functional API
evaluate_model(best)
```

```python
# OOP API
s.evaluate_model(best)
```

# Make Predictions

```python
predictions = predict_model(best, data = test)
predictions.head(10)
```

```python
predictions[predictions['prediction_score'] < .9]
```
